```python
# This notebook is accompanied by a Medium article: 
# https://python.plainenglish.io/starting-with-tensorflow-on-kaggle-0785fb83c86d
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

num_neurons = 512
num_classes = 10
epochs = 10
```

```python
train_data = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
test_data = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
```

```python
print(train_data.shape)
```

```python
x_train=train_data.drop('label',axis=1).values.reshape(len(train_data),28,28,1)
y_train=train_data.label.values
x_test = test_data.values.reshape(len(test_data),28,28,1)
```

```python
print(x_train.shape)
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = Sequential([Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)),
                    MaxPooling2D(2,2),
                    Conv2D(32, (3,3), activation='relu'),
                    MaxPooling2D(2,2),
                    Flatten(),
                    Dropout(0.5),
                    Dense(num_neurons, activation=tf.nn.relu), 
                    Dense(num_classes, activation=tf.nn.softmax)])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=epochs)
```

```python
y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred,axis=1)
```

```python
submit = pd.read_csv('/kaggle/input/digit-recognizer/sample_submission.csv')
submit.Label = y_pred
submit.to_csv('submission.csv',index=False)
```
